````python
import os
import numpy as np
import matplotlib.pyplot as plt

# twissed
import twissed

# Select colormaps
cm = twissed.Cmap()
````

````python
# Selection of the directory with data
directory = os.getcwd() + '/data/FBPIC/lab_diags'

# Find all timesteps
steps = twissed.Steps()
steps.find_data(directory=directory,source='fbpic',verbose=False)

# timestep selection
timestep = steps.timesteps[-1]

#Creation of the step class
step = twissed.Step()

# Read data 
step = twissed.fbpic_particle(step,directory,timestep,species='electrons',verbose=False)

_ = step.hist2D('z','Ek',xconv='um',range_auto=True)
````

````python
with plt.rc_context(twissed.rcParams):
    fig, ax = plt.subplots(figsize=(6,6), dpi=100, tight_layout=True)

H, xedges = step.hist1D('Ek',bins=20,plot='plot',ax=ax,fwhm=False)

H, xedges = step.hist1D('Ek',bins=30,plot='plot',ax=ax,fwhm=False)

H, xedges = step.hist1D('Ek',bins=40,plot='plot',ax=ax,fwhm=False)

H, xedges = step.hist1D('Ek',bins=50,plot='plot',ax=ax,fwhm=False)

H, xedges = step.hist1D('Ek',bins=60,plot='plot',ax=ax,fwhm=False)

H, xedges = step.hist1D('Ek',bins=70,plot='plot',ax=ax,fwhm=False)
````
